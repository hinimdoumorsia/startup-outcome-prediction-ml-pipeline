#### 2.b. Delete outliers for 'funding_total_usd' and 'funding_rounds'.

```python
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
company['funding_total_usd'] = pd.to_numeric(company['funding_total_usd'], errors='coerce')
company['funding_rounds'] = pd.to_numeric(company['funding_rounds'], errors='coerce')
company = remove_outliers_iqr(company, 'funding_total_usd')
company = remove_outliers_iqr(company, 'funding_rounds')
print(f"Shape after removing outliers: {company.shape}")
```

# **Interpretation of Active Duration Calculation and Contradiction Handling**

- Calculates the active duration of a company by subtracting **`founded_at`** from **`closed_at`**, resulting in a **`timedelta`** stored in **`active_years`**.
- Converts this duration to an approximate number of days by multiplying by 365, stored in **`active_days`**.
- Detects contradictions where **`active_days`** is negative (i.e., **`closed_at`** is before **`founded_at`**), which are logically inconsistent.
- Replaces these negative durations with **`NaN`** to mark invalid data.
- Prints the count of such contradictory records.
- Displays a sample of contradictory rows showing founding year, closing date, duration fields, and company status for inspection.

This process cleans the data by identifying and handling impossible active durations.


```python
print(company['active_days'].unique())
```

```python
import numpy as np
from scipy.stats import shapiro
active_days_non_nan = company['active_days'].dropna()
mean_val = active_days_non_nan.mean()
median_val = active_days_non_nan.median()
stat, p_value = shapiro(active_days_non_nan)
print(f"p-value test normalité active_days (sans NaN) : {p_value:.4f}")
if p_value > 0.05:
    fill_value = mean_val
    print(f"Distribution normale → remplissage avec la moyenne : {fill_value:.2f}")
else:
    fill_value = median_val
    print(f"Distribution non normale → remplissage avec la médiane : {fill_value:.2f}")
company['active_days'] = company['active_days'].fillna(fill_value)
print("Valeurs uniques dans active_days après remplissage :")
print(company['active_days'].unique())
```

# **Interpretation of Active Days Imputation**

- The code selects non-missing values from the **`active_days`** column.
- It calculates the mean and median of these values.
- Performs the Shapiro-Wilk test to check if **`active_days`** follows a normal distribution.
- Based on the p-value:
  - If data is normally distributed **(`p > 0.05`)**, missing values are filled with the mean.
  - Otherwise, missing values are filled with the median.
- Finally, it fills the missing values in **`active_days`** accordingly and prints the unique values after imputation.

This approach ensures an appropriate imputation method based on the distribution of the data.

# **DataFrame Cleanliness Check**

This Python script analyzes a DataFrame `company` to identify potential data issues:

- **Missing values** (`NaN`, `None`) per column, with their count and percentage.  
- **Null-like or empty values**, such as empty strings `""` or textual strings `"NA"`, `"null"`, `"nan"`.  
- **Outliers detected using the IQR method** (interquartile range) across all numeric columns, counting the number of outliers per column.  
- A global summary showing the total number of rows, columns, and the count of columns affected by each of the above issues.


```python
import pandas as pd
import numpy as np

def check_dataframe_cleanliness(df):
    total_rows = df.shape[0]
    print("=== Valeurs manquantes (NaN, None) ===")
    missing = df.isnull().sum()
    missing_percent = 100 * missing / total_rows
    missing_summary = pd.DataFrame({'missing_count': missing, 'missing_percent': missing_percent})
    print(missing_summary[missing_summary['missing_count'] > 0].sort_values('missing_percent', ascending=False))
    print()
    print("=== Valeurs nulles ou vides spécifiques ('', 'NA', 'null') ===")
    def count_null_like(col):
        return col.astype(str).isin(['', 'NA', 'null', 'nan']).sum()
    null_like = df.apply(count_null_like)
    null_like_percent = 100 * null_like / total_rows
    null_like_summary = pd.DataFrame({'null_like_count': null_like, 'null_like_percent': null_like_percent})
    print(null_like_summary[null_like_summary['null_like_count'] > 0].sort_values('null_like_percent', ascending=False))
    print()
    print("=== Outliers détectés par méthode IQR (colonnes numériques) ===")
    num_cols = df.select_dtypes(include=[np.number]).columns
    outlier_counts = {}
    for col in num_cols:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)][col]
        outlier_counts[col] = outliers.count()
    outlier_df = pd.DataFrame.from_dict(outlier_counts, orient='index', columns=['outlier_count'])
    outlier_df = outlier_df[outlier_df['outlier_count'] > 0].sort_values('outlier_count', ascending=False)
    print(outlier_df)
    print()
    print("=== Résumé global ===")
    print(f"Total lignes : {total_rows}")
    print(f"Total colonnes : {df.shape[1]}")
    print(f"Colonnes avec NaN : {len(missing_summary[missing_summary['missing_count'] > 0])}")
    print(f"Colonnes avec valeurs nulles/vide : {len(null_like_summary[null_like_summary['null_like_count'] > 0])}")
    print(f"Colonnes avec outliers détectés : {len(outlier_df)}")
check_dataframe_cleanliness(company)
```

# **Data Cleanliness Report Summary**

- **Missing values**: Only 2 columns (`first_funding_days`, `last_funding_days`) have about 0.32% missing data.  
- **Null/empty values**: These same columns also contain empty or null-like string values that need cleaning.  
- **Outliers**: 66 out of 69 columns contain extreme values detected by the IQR method, some with several thousand outliers.  
- **Conclusion**: Missing values should be imputed and outliers treated to ensure data quality before analysis or modeling.

```python
null_like_values = ['', 'NA', 'null', 'nan']
company = company.replace(null_like_values, np.nan)
for col in ['first_funding_days', 'last_funding_days']:
    if col in company.columns:
        mean_val = company[col].mean()
        company[col].fillna(mean_val, inplace=True)
num_cols = company.select_dtypes(include=[np.number]).columns
for col in num_cols:
    Q1 = company[col].quantile(0.25)
    Q3 = company[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    company[col] = np.where(company[col] < lower_bound, lower_bound, company[col])
    company[col] = np.where(company[col] > upper_bound, upper_bound, company[col])
```
