## ✅ Missing Value Treatment: Methodological Approach

###  Objective
Fill missing values in numerical columns in a statistically sound way to avoid introducing bias into the model.

---

### Approach Steps:

#### 1. Identify Numerical Columns
Determine which columns contain numerical variables (e.g., funding amounts, number of rounds, ROI, etc.).

---

#### 2. Analyze the Distribution
For each numerical column:
- Assess the **shape of the distribution** using statistical indicators such as **skewness**.
- A **symmetric distribution** suggests the mean is a representative measure.
- An **asymmetric distribution** (long tail on one side) likely indicates the presence of **outliers**.

---

#### 3. Choose the Imputation Method
- If the variable is **symmetric (skew ≈ 0)**: fill missing values with the **mean**.
- If the variable is **asymmetric (skew > 1 or skew < -1)**: use the **median**, which is more robust to outliers.

---

#### 4. Replace Missing Values
Apply the chosen imputation for each column according to its distribution:
- Mean if the distribution is normal/symmetric
- Median if the distribution is skewed/asymmetric

---

#### 5. Final Verification
Ensure all missing values have been addressed:
- No cells should contain `NaN`.
- Descriptive statistics should be consistent with expectations.

```python
numerical_columns = [
    'investment_rounds',
    'invested_companies',
    'funding_rounds',
    'funding_total_usd',
    'milestones',
    'ROI'
]
categorical_columns = [
    'name',
    'category_code',
    'status',
    'short_description',
    'description',
    'country_code',
    'state_code',
    'city',
    'region',
    'domain_extension'
]
df_cleaned[numerical_columns].skew()
for col in numerical_columns:
    if abs(df_cleaned[col].skew()) > 1:
        df_cleaned[col].fillna(df_cleaned[col].median(), inplace=True)
    else:
        df_cleaned[col].fillna(df_cleaned[col].mean(), inplace=True)
print("\n✅ Colonnes numériques sans NaN :")
for col in numerical_columns:
    if df_cleaned[col].isnull().sum() == 0:
        print(f"{col} : OK")
    else:
        print(f"{col} : ❌ {df_cleaned[col].isnull().sum()} NaNs restants")
```

# **Summary of Steps for Processing Datetime Columns**

## **1. Calculate the median of each datetime column**

- Sort the non-missing dates and take the median value.  
- **Why?**  
  The median is robust to extreme values, unlike the mean.

## **2. Replace missing values (NaN) with the median date**

- This avoids dropping rows and limits data distortion.  
- **Why?**  
  Replacing with the median maintains temporal consistency without introducing extreme bias.

## **3. Calculate the difference in days for each pair of related columns**

- For example, the difference between `first_investment_at` and `last_investment_at`.  
- Create a new column that contains this duration in days.  
- **Why?**  
  This converts temporal information into numerical values that can be used for analysis or machine learning.


```python
datetime_cols = [
    'founded_at', 'closed_at',
    'first_investment_at', 'last_investment_at',
    'first_funding_at', 'last_funding_at',
    'first_milestone_at', 'last_milestone_at'
]
# 1. Convertir en datetime si ce n’est pas encore fait
for col in datetime_cols:
    df_cleaned[col] = pd.to_datetime(df_cleaned[col], errors='coerce')
# 2. Remplir les valeurs manquantes par la médiane
for col in datetime_cols:
    median_date = df_cleaned[col].dropna().sort_values().iloc[len(df_cleaned[col].dropna()) // 2]
    df_cleaned[col].fillna(median_date, inplace=True)
# 3. Créer les colonnes avec différence en jours avec noms complets
date_pairs = [
    ('founded_at', 'closed_at'),
    ('first_investment_at', 'last_investment_at'),
    ('first_funding_at', 'last_funding_at'),
    ('first_milestone_at', 'last_milestone_at')
]
diff_columns = []
for start_col, end_col in date_pairs:
    # Nom de colonne basé sur la paire complète
    suffix = start_col.replace('first_', 'first_').replace('_at', '')  # récupère le nom logique
    diff_col_name = f"days_first_last_{suffix}_at"
    df_cleaned[diff_col_name] = (df_cleaned[end_col] - df_cleaned[start_col]).dt.days
    diff_columns.append(diff_col_name)
# 4. Supprimer les colonnes de date utilisées pour le calcul
cols_to_drop = set(sum(date_pairs, ()))
df_cleaned.drop(columns=list(cols_to_drop), inplace=True)
# Création de la variable cible binaire
df_cleaned['target'] = df_cleaned['status'].apply(lambda x: 1 if x in ['Operating', 'IPO'] else 0)
df_cleaned = df_cleaned.drop(columns=['status'])
```

## **Encoding Categorical Variables: One-Hot and Label Encoding Techniques**

## Frequency Encoding of Company Names

## What is it?

Frequency Encoding replaces each company name with its relative frequency of occurrence in the dataset.

---

## How does it work?

For each name, we calculate:
$$
\text{Encoded value} = \frac{\text{Number of occurrences of the name}}{\text{Total number of rows in the dataset}}
$$

---

## Why use it?

- Transforms a categorical variable with **many unique values** into a continuous numerical variable.
- Avoids creating thousands of columns (as with one-hot encoding).
- Allows models like SVM and Random Forest to easily handle this variable.
- Reflects the relative popularity of a name, which can be informative.

---



```python
# List of categorical columns to frequency encode
cols_to_encode = ['name', 'category_code', 'country_code', 'state_code', 'region', 'city']
# Apply frequency encoding to each column
for col in cols_to_encode:
    freq = df_cleaned[col].value_counts(normalize=True)  # Calculate relative frequencies
    df_cleaned[f'{col}_freq_enc'] = df_cleaned[col].map(freq).fillna(0)  # Map frequencies and handle NaNs
# Drop the original columns after encoding
df_cleaned.drop(columns=cols_to_encode, inplace=True)
```

### Mathematical description of the text processing

- Let $D = \{d_1, d_2, \ldots, d_N\}$ be the set of long descriptions, and $S = \{s_1, s_2, \ldots, s_N\}$ be the set of short descriptions.

- $V_D = \{w_1, w_2, \ldots, w_{M_D}\}$ is the vocabulary of unique words in $D$, and $V_S = \{u_1, u_2, \ldots, u_{M_S}\}$ is the vocabulary in $S$.

- Relative frequency of a word $w \in V_D$:
  
  $$
  f_D(w) = \frac{\text{number of occurrences of } w \text{ in } D}{\sum_{w' \in V_D} \text{number of occurrences of } w'}
  $$

- Relative frequency of a word $u \in V_S$:
  
  $$
  f_S(u) = \frac{\text{number of occurrences of } u \text{ in } S}{\sum_{u' \in V_S} \text{number of occurrences of } u'}
  $$

- For the long description $d_i$ containing words $\{w_{i1}, \ldots, w_{iK_i}\}$, compute:
  
  $$
  \text{desc\_freq\_sum}_i = \sum_{k=1}^{K_i} f_D(w_{ik})
  $$

- For the short description $s_i$ containing words $\{u_{i1}, \ldots, u_{iL_i}\}$, compute:
  
  $$
  \text{short\_desc\_freq\_sum}_i = \sum_{l=1}^{L_i} f_S(u_{il})
  $$

- These two values are used as compact numerical features for each row in the dataset.

```python

from collections import Counter
# Description longue : description
df_cleaned['desc_clean'] = df_cleaned['description'].fillna('').str.lower()
desc_words = ' '.join(df_cleaned['desc_clean']).split()
desc_freq = Counter(desc_words)
total_desc_words = sum(desc_freq.values())
for word in desc_freq:
    desc_freq[word] /= total_desc_words
def sum_desc_freq(text):
    return sum(desc_freq.get(word, 0) for word in text.split())
df_cleaned['desc_freq_sum'] = df_cleaned['desc_clean'].apply(sum_desc_freq)
df_cleaned.drop(columns=['desc_clean'], inplace=True)
# Description courte : short_description
df_cleaned['short_desc_clean'] = df_cleaned['short_description'].fillna('').str.lower()
short_words = ' '.join(df_cleaned['short_desc_clean']).split()
short_freq = Counter(short_words)
total_short_words = sum(short_freq.values())
for word in short_freq:
    short_freq[word] /= total_short_words
def sum_short_desc_freq(text):
    return sum(short_freq.get(word, 0) for word in text.split())
df_cleaned['short_desc_freq_sum'] = df_cleaned['short_desc_clean'].apply(sum_short_desc_freq)
df_cleaned.drop(columns=['short_desc_clean'], inplace=True)
df_cleaned.drop(columns=['description', 'short_description'], inplace=True)
```
